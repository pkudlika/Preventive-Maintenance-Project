```python
#Drop scanner columns 
df2=df1.drop('scanner',axis=1)
#Order the columns accordingly
df2[['date','min','max','pixels','err']]
```

```python
#Checking the relationship of err column by using formula 
df2['theoretical_err']=((df2['max'].astype(float) \
            .sub(df2['min'].astype(float), axis=0)) \
            .div(df2['min'].astype(float),axis=0))\
            .mul(float(100))
df2
```

```python
#Normalizing the input variables
from sklearn.preprocessing import Normalizer

df2_n = Normalizer().fit_transform(df2)
df2_n = pd.DataFrame(df2_n)
```

```python
#Assigning the input variable x 
x = df2_n.iloc[:,2:5]
x
```

```python
#Target variable
y = df2['err']
y
```

```python
#Import linear regression model
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
#Fitting the model
regr.fit(x,y)
```

```python
#Predicting 
y_pred = regr.predict(x)
pd.DataFrame(y_pred)
```

```python
#Assigning a new dataframe to target 
c = pd.DataFrame(y_pred)
c.shape
```

```python
df2.shape
```

```python
#Checking the null values 
c = pd.DataFrame(y_pred)
c.isnull().sum()
```

```python
#Joining the previous dataframe with c
df_new = df2.join(c)
df_new.isnull().sum()
```

```python
#Renaming the columns
df4 = df_new.rename(columns={0 : 'err_pred'})
df4
```

```python
#Setting the threshold as 12 (survival gap)
df4['survival_gap'] = 12 - df4['err_pred']
df_final = df4 
```

```python
df_final
```

```python
#Ordering the column names
df_final[['date','min','max','pixels','theoretical_err','err','err_pred','survival_gap']]
```

```python
#Checking the err_pred > 12 values 
df_final.loc[(df_final['err_pred']>12.0)]
```

```python
#Checking err_pred<=12
df_final.loc[(df_final['err_pred']<=12.0)]
```

```python
df_final.loc[df_final['survival_gap'] <=0]
```

```python
#Creating a function to determine the scanner is in good condition or not 
def message(df_final):
    if (df_final['survival_gap'] <= 0) :
        return 'Scanner is damaged'
    elif(0 < df_final['survival_gap'] < 2) :
        return 'Scanner is about to damage'
    else :
        return 'Scanner is good'
df_final['message'] = df_final.apply(message, axis = 1)    
df_final
```
